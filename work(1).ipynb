```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(6.4*5, 4.8*5), constrained_layout=False)

img_c1 = cv2.imread("period_input.jpg", 0)
H = np.ones(img_c1.shape[:2])
H[110:130,310:330]=0
H[230:250,480:500]=0
H[230:250,140:160]=0
H[350:370,310:330]=0
H[160:180,230:250]=0
H[290:310,230:250]=0
H[290:310,400:420]=0
H[160:180,400:420]=0
img_c2 = np.fft.fft2(img_c1)
img_c3 = np.fft.fftshift(img_c2)
img_c6 = H*np.fft.fftshift(img_c2)
img_c4 = np.fft.ifftshift(img_c6)
img_c5 = np.fft.ifft2(img_c4)
```

```python
plt.subplot(141), plt.imshow(img_c1, "gray"), plt.title("Original Image")
plt.subplot(142), plt.imshow(np.log(1+np.abs(img_c2)), "gray"), plt.title("Spectrum")
plt.subplot(143), plt.imshow(np.log(1+np.abs(img_c3)), "gray"), plt.title("Centered Spectrum")
plt.subplot(144), plt.imshow(H, "gray"), plt.title("Designed Filter")
plt.show()
```

```python
plt.subplot(131), plt.imshow(np.log(1+np.abs(img_c6)), "gray"), plt.title("H*Centered Spectrum")
plt.subplot(132), plt.imshow(np.log(1+np.abs(img_c4)), "gray"), plt.title("Decentralized")
plt.subplot(133), plt.imshow(np.abs(img_c5), "gray"), plt.title("Processed Image")

plt.show()
```
